```python
import pandas as pd

df = pd.read_csv("/Users/keerti/Library/Mobile Documents/com~apple~Numbers/Documents/students_data.csv")

print("DataFrame:")
print(df)

print("\nFirst Five Records:")
print(df.head())

print("\nLast Five Records:")
print(df.tail())

print("\nTotal Number of Rows and Columns:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nDataset Information:")
df.info()
```

```python
print("Attendance Mean:")
print(df["Attendance"].mean())

print("\nAverage Internal Marks:")
print(df["Internal marks"].mean())

print("\nAverage External Marks:")
print(df["External marks"].mean())

print("\nMaximum Internal Marks:")
print(df["Internal marks"].max())

print("Minimum Internal Marks:")
print(df["Internal marks"].min())

print("\nMaximum External Marks:")
print(df["External marks"].max())

print("Minimum External Marks:")
print(df["External marks"].min())

print("\nStandard Deviation of Internal Marks:")
print(df["Internal marks"].std())

print("\nSummary Statistics:")
print(df.describe())
```

```python
df["Total Marks"] = df["Internal marks"]+df["External marks"]
df["Percentage"] = (df["Total Marks"]/100)*100
print(df)
```

```python
print("Students with more than 90% attendance: ")
print(df[df["Attendance"]>90])
print("Students with more than 90 Marks: ")
print(df[df["Total Marks"]>90])
print("Feamle Students: ")
print(df[df["Gender"]=="F"])
print("Male Students: ")
print(df[df["Gender"]=="M"])
print("Students with less than 50 extrenal marks: ")
print(df[df["External marks"]<50])
```

```python
print("Highest Total Marks: ")
print(df.loc[df["Total Marks"].idxmax()])
print("Lowest Total Marks: ")
print(df.loc[df["Total Marks"].idxmin()])
print("Top Three Preformers: ")
print(df.sort_values(by="Total Marks", ascending=False).head(3))
print("Bottom Three Preformers: ")
print(df.sort_values(by="Total Marks", ascending=True).head(3))
class_average = df["Total Marks"].mean()
above_average = df[df["Total Marks"] > class_average]
print("Number of above average Students: ",len(above_average))
```

```python
print("AVERAGE TOTAL MARKS OF MALE STUDENTS")
print(df[df['Gender'] == 'M']['Total Marks'].mean())
print("AVERAGE TOTAL MARKS OF FEMALE STUDENTS")
print(df[df['Gender'] == 'F']['Total Marks'].mean())
print("AVERAGE ATTENDANCE BY GENDER")
print(df.groupby('Gender')['Attendance'].mean())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.bar(df['Name'], df['Total Marks'])
plt.title("Total Marks of Students")
plt.xlabel("Students")
plt.ylabel("Total Marks")
plt.show()
plt.figure(figsize=(8,5))
plt.hist(df['Attendance'], bins=5)
plt.title("Attendance Distribution")
plt.xlabel("Attendance")
plt.ylabel("Frequency")
plt.show()
gender_count = df['Gender'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gender_count,
        labels=gender_count.index,
        autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.show()
plt.figure(figsize=(8,5))
plt.scatter(df['Attendance'], df['Total Marks'])
plt.title("Attendance vs Total Marks")
plt.xlabel("Attendance")
plt.ylabel("Total Marks")
plt.show()
```
